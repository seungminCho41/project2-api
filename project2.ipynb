## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#save and print the text content of a page with all tags removed

source = requests.get('https://en.wikipedia.org/wiki/Detroit:_Become_Human').text
soup = BeautifulSoup(source) 
print(soup.get_text())
```

```python
#download an image with beautifulsoup and save it in this repository

img_tags=soup.find_all('img') 
img_url='https:' + img_tags[5]['src']
img=requests.get(img_url)
with open('image.png','wb') as f:
    f.write(img.content)
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links

links = soup.find_all('a') 
for i in range(1, 10):
    print(links[i].get('href')[:100])
```

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 

api_key = '4e4af9fd949ab6a91aa714d4cf669373'

url = f"https://api.themoviedb.org/3/movie/now_playing?api_key={api_key}"

response = requests.get(url).json()
movies = response['results']

# Extract only the required metadata fields
movies_metadata = [{'title': movie['title'], 'release_date': movie['release_date'], 'rating': movie['vote_average'], 'overview': movie['overview']} for movie in movies]

movies_df = pd.DataFrame(movies_metadata)
movies_df.to_csv('movies.csv', index=False)
movies_df

#save the movie metadata as a csv file
movies_df.to_csv('movies.csv', index=False)
movies_df
```

```python
#download the movie posters for 10 of these movies and save them to this repository

def download_poster(movie):
    try:
        poster_path = movie['poster_path']
        poster_url = f"https://image.tmdb.org/t/p/original{poster_path}"
        response = requests.get(poster_url)
        response.raise_for_status()
        with open(f"{movie['title']}.jpg", "wb") as f:
            f.write(response.content)
    except requests.exceptions.RequestException as e:
        print(f"Error fetching poster for {movie['title']}: {e}")

# Download posters for the first 10 movies
for movie in movies[:10]:
    download_poster(movie)

```
